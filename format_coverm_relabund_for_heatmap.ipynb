## This notebook formats the output of CoverM genome relative abundance calculations for the heatmap.py script

#### Import all required modules

```python
import pandas as pd
import os as os
import glob as glob
import numpy as np
```

#### Navigate to directory with CoverM outputs

```python
os.chdir("/Users/nastassia.patin/GitHub/eDNA-Visualizations/MAG_relativeabundances")
```

##### Import sheet and truncate metagenome sample names (column names)

```python
# Flyer 2018
df = pd.read_csv("Flyer2018_MAG_relabund.tsv", sep='\t')
df.columns = df.columns.str.replace("eCruises_2018-2021_MAGs_HQ_dRep_concat.fa/", 
                                  "", regex=False)
df.columns = df.columns.str.replace("R1_interleaved.fq.gz Relative Abundance (%)", 
                                  "", regex=False)
df.head(3)
```

```python
# Lasker 2018
df = pd.read_csv("Lasker2018_MAG_relabund.tsv", sep='\t')
df.columns = df.columns.str.replace("eCruises_2018-2021_MAGs_HQ_dRep_concat.fa/", 
                                  "", regex=False)
df.columns = df.columns.str.replace("_S[0-9]+_L002_interleaved.fq.gz ", 
                                  "", regex=True)
df.columns = df.columns.str.replace("Relative Abundance (%)", 
                                  "", regex=False)
df.head(3)
```

```python
# Lasker 2019
df = pd.read_csv("Lasker2019_MAG_relabund.tsv", sep='\t')
df.columns = df.columns.str.replace("eCruises_2018-2021_MAGs_", 
                                  "", regex=False)
df.columns = df.columns.str.replace("-sorted Relative Abundance (%)", 
                                  "", regex=False)
df.head(3)
```

```python
# Lasker 2021
df = pd.read_csv("Lasker2021_MAG_relabund.tsv", sep='\t')
df.columns = df.columns.str.replace("eCruises_2018-2021_MAGs_HQ_dRep_concat.fa/", 
                                  "", regex=False)
df.columns = df.columns.str.replace("_interleaved.fq.gz Relative Abundance (%)", 
                                  "", regex=False)
df.head(3)
```

##### Remove the 'unmapped' row

```python
df = df.iloc[1:, :]
df.head(3)
```

##### Remove all MAGs generated from control samples (Flyer 2018)

```python
controls = ['zr2760-13', 'zr2760_34', 'zr2760-8', 'zr2760-9']
df = df.drop(df[df['Genome'].str.contains('|'.join(controls))].index)
df.tail(3)
```

##### Remove all control samples (Flyer 2018 only)

```python
controls = ['zr2760_1_', 'zr2760_5_', 'zr2760_8_', 'zr2760_13', 'zr2760_22',
            'zr2760_34', 'zr2760_72', 'zr2760_80', 'zr2760_81', 'zr2760_82', 
            'zr2760_83', 'zr2760_88', 'zr2760_90', 'zr2760_91', 'zr2760_92', 
            'zr2760_93', 'zr2760_94', 'zr2760_95']
```

```python
df = df.loc[:, ~df.columns.str.contains('|'.join(controls))]
df.head(3)
```

### Merge the genome taxonomy level of interest with relative abundance sheet

#### Navigate to directory containing MAG taxonomies

```python
os.chdir("/Users/nastassia.patin/Desktop/Projects/eCruises_TIMESERIES/Metagenomes_2018-2021/MAG_QC/")
```

##### Import the MAG taxonomy sheet

```python
tax = pd.read_csv("MAGs_2018-2021_HQ_dRep_quality_tax_noCs.csv")
tax.head(3)
```

```python
taxon = 'phylum'
tax = tax[['bin', taxon]].rename(columns={'bin':'Genome'})
tax['Genome'] = tax['Genome'].str.replace('_', '-')
tax.head(3)
```

```python
df_tax = df.merge(tax, on='Genome', how='left')
df_tax.head(3)
```

```python
# get a list of columns
cols = list(df_tax)
# move the column to head of list using index, pop and insert
cols.insert(0, cols.pop(cols.index(taxon)))
df_tax = df_tax.loc[:, cols]
df_tax.head(3)
```

#### Remove all MAGs in the phylum Firmicutes

```python
df_tax = df_tax.drop(df_tax[df_tax['phylum'].str.contains('Firmicutes')].index)
df_tax.head(3)
```

```python
df_tax = df_tax.dropna(axis='columns')
```

##### Export

```python
os.chdir("/Users/nastassia.patin/GitHub/eDNA-Visualizations/MAG_relativeabundances")
df_tax.to_csv("Lasker2021_MAG_relabund_forheatmap-phylum.tsv", sep='\t', index=False)
```

#### Log-normalize and export

```python
df_tax = df_tax.set_index([taxon, 'Genome'])
df_tax = df_tax + 0.000000001
df_tax = np.log(df_tax)
df_tax.head(3)
```

```python
df_tax.to_csv("Lasker2021_MAG_relabund_log_forheatmap-phylum.tsv", 
              sep='\t', index=True)
```
