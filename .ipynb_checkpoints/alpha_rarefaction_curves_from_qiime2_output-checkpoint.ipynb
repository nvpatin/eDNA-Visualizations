```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os as os
```

```python
os.chdir('/Users/nastassia.patin/Desktop/Projects/NOAA eDNA/Review paper/alpha diversity/eCruises-Padilla-Hallam')
inf = 'observed_features.csv'
otus = pd.read_csv(inf, index_col=0, sep=',')
```

```python
cols = [col for col in otus.columns if 'iter' not in col]
mean, data = otus[cols], pd.DataFrame(columns=cols)
depths = [col.split('_')[0] for col in otus.columns if 'depth' in col]
otus = otus.drop(cols, axis=1)
otus.columns = depths
for depth in depths:
    mean['ASV'] = otus[depth].mean(axis=1)
    mean['depth']= depth.split('-')[-1]
    data = pd.concat([data,mean])
```

```python
data
```

```python
plt.figure(figsize=(14,8))
sns.set(font_scale = 1.5)
sns.set_style("white")

# use your metadata column name of choice to plot hue and style
curves = sns.lineplot(x='depth', y='ASV', data=data, palette="colorblind", 
                      hue='volume_filtered_mL', sort=False, ci='sd', err_style=None)
plt.xlabel('Sequencing depth', fontsize=16)
plt.ylabel('Observed OTUs/ASVs', fontsize=16)
plt.xticks(size=14, rotation=45)

handles, labels = curves.get_legend_handles_labels()
plt.legend(handles=handles[0:], labels=labels[0:], prop={'size':14}, frameon=True, 
           borderaxespad=0, handletextpad=1.0, loc='upper left', title="Volume Filtered (mL)",
          bbox_to_anchor=(1.02, 1))
plt.savefig('Observed_ASVs.svg', format='svg', dpi=1200, bbox_inches='tight')
```
